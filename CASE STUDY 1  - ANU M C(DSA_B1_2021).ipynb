# 3. Plot a bar graph of male vs female buyers participated in the sales.

# 4. Find the top 5 cars based on their sales price.

```python
d2 = data1.sort_values(by = ['Sale Price'], ascending=False)
d3 = d2.loc[:, ["Make", "Model", "Sale Price"]]
d3.head()
```

```python
c1 = data1.nlargest(5, 'Sale Price')
c1.loc[:, ["Make", "Model", "Sale Price"]]
```

# Insights

- Volkswagen Jetts III is the most expensive car in the list with a price tag of 99997.65.
-Audi A6, Lexus LX, Mazda MX_6,Lexus ES are the followers in the list with a price tag of 99991.91, 99983.38,99978.17,99973.44 respectively.
-Two Lexus models has place in most five expensive cars.
